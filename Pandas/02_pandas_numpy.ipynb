# What is Pandas
Pandas is a open source data analysis library written in python<br>
It liverages the power and speed of numpy to make data analysis and processing easy for data scientists<br>
It provides rich and highly robust data operations.


###  Two types of data structures in Pandas
- a) Series --> It's a one dimensional array with indexes, it stores a single column or row of data in a Dataframe

- b) Dataframe --> It's a tabular spreadsheet like structure representing rows each of which contains one or multiple columns

**A one dimensional array(labeled) capable of holding any type of data-Series**

**A two-dimensional array(labeled) structure with columns of potentially different types of data- DataFrame**

# iloc


```python
# using loc, we can select row/column using it's index name 

# so imagine that my rows and column names might be different
# and selecting them based on their index names is difficult
# to acces any row.column/data using indexing(not real names) then use iloc

df.iloc[:,1] # to select first column
df.iloc[0,:] # to select zeroth row
df.iloc[1,2] # to select fisrt row second column value
df.iloc[[1,5],[1,5]] # to select 1st and 5nd rows and columns
```

# Removing Duplicates


```python
# 1. Remove All Duplicate Rows
df_uniq = df.drop_duplicates()

# 2. Remove Duplicates Based on Specific Columns
df_uniq = df.drop_duplicates(subset=['A'])


# 3. Keep Last Occurrence Instead of First
df_uniq = df.drop_duplicates(keep='last')

# 4. Remove All Duplicates (No First or Last)
df_uniq = df.drop_duplicates(keep=False)

# 5. Remove Duplicates In-Place
# To modify the DataFrame without creating a new one:
df_uniq = df.drop_duplicates(inplace=True)
```

# Removing Nulls

```python
# 1. Remove Rows with Any Null Value
df_new = df.dropna()

# 2. Remove Rows Only If All Values Are NaN
df_new = df.dropna(how='all')

# 3. Remove Rows Based on a Specific Column
df_new = df.dropna(subset='A')



# 4. Remove Columns with Any Null Values
df_new = df.dropna(axis=1)

# 5. Fill Missing Values Instead of Removing
df_new = df.fillna(0)


```

# df.info

The df.info() function provides a summary of a Pandas DataFrame, including:

1. Number of entries (rows) and columns
2. Column names and data types
3. Number of non-null (non-NaN) values in each column
4. Memory usage of the DataFrame
